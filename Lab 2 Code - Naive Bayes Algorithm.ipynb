```python
# Assigning features and label variables (data was faked)
# Two features: outlook and temprature
outlook=['Sunny','Sunny','Overcast','Rainy','Rainy','Rainy','Overcast','Sunny','Sunny',
'Rainy','Sunny','Overcast','Overcast','Rainy']
temp=['Hot','Hot','Hot','Mild','Cool','Cool','Cool','Mild','Cool','Mild','Mild','Mild','Hot','Mild']
# Label (play or not)
play=['No','No','Yes','Yes','Yes','No','Yes','No','Yes','Yes','Yes','Yes','Yes','No']
# What is this problem?
```

```python
# Import LabelEncoder
# A LabelEncoder is a utility class commonly used in machine learning and data preprocessing to transform categorical labels into numerical values. It is particularly useful when working with algorithms that require numerical inputs, such as many machine learning models. A LabelEncoder assigns a unique integer to each category or label in a categorical feature, thereby converting it into a numerical format.
from sklearn import preprocessing
#creating labelEncoder
le = preprocessing.LabelEncoder()
# Converting string labels into numbers
outlook_encoded=le.fit_transform(outlook)
print('Outlook', outlook_encoded)
```

```python
# Converting string labels into numbers
temp_encoded=le.fit_transform(temp)
label=le.fit_transform(play)
print ('Temp:',temp_encoded)
print ('Play:',label)
```

```python
#Combinig outlook and temp into single list of tuples
features=list(zip(outlook_encoded,temp_encoded))
print (features)
```

```python
#Import Gaussian Naive Bayes model
from sklearn.naive_bayes import GaussianNB

#Create a Gaussian Classifier
model = GaussianNB()
# We didn't split the data because we have a limited amount of it
# Train the model using the training sets
model.fit(features,label)

```

```python
# Predict Output
predicted= model.predict([[0,2]]) # 0:Overcast, 2:Mild
if (predicted == 1):
   print ("Predicted Value:", 'Yes')
else:
   print ("Predicted Value:", 'No')

# Use another set of data
```

Task 1: Use Lab 2 dataset to run the Naive Bayes algorithm using a CSV file.

```python
import pandas as pd
data = pd.read_csv('datasets/Lab 2 dataset.csv')

# EDA
data.head()
```

```python
data.info()
```

```python
label_encoder = preprocessing.LabelEncoder()
data['Outlook'] = label_encoder.fit_transform(data['Outlook'])  
data['Temperature'] = label_encoder.fit_transform(data['Temperature'])
data['Humidity'] = label_encoder.fit_transform(data['Humidity'])
data['Windy'] = label_encoder.fit_transform(data['Windy'])
data['Play Golf'] = label_encoder.fit_transform(data['Play Golf'])

features = data[['Outlook', 'Temperature', 'Humidity', 'Windy']]

model1 = GaussianNB()
model1.fit(features, data['Play Golf'])

predicted1 = model1.predict([[0, 2, 0, 1]]) # 0:Overcast, 2:Mild, 0:High, 1:True
if (predicted1 == 1):
   print ("Predicted Value:", 'Yes')
else:
    print ("Predicted Value:", 'No')
```

```python
# model accuracy etc
f1 = model1.score(features, data['Play Golf'])
f1.round(2)
```
